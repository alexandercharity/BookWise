# BookWise — EDA & Preprocessing
Dataset: GoodBooks-10k (Kaggle)
https://www.kaggle.com/datasets/zygmuntz/goodbooks-10k

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DATA_DIR = '../../data'

# GoodBooks-10k: books.csv & ratings.csv
books   = pd.read_csv(f'{DATA_DIR}/books.csv')
ratings = pd.read_csv(f'{DATA_DIR}/ratings.csv')

print('Books shape:  ', books.shape)
print('Ratings shape:', ratings.shape)
print('\nBooks columns:', books.columns.tolist())
print('Ratings columns:', ratings.columns.tolist())
```

```python
# Preview data
print('=== Books Sample ===')
print(books[['book_id','title','authors','original_publication_year','average_rating']].head())
print('\n=== Ratings Sample ===')
print(ratings.head())
```

```python
# Distribusi rating (skala 1-5)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

ratings['rating'].value_counts().sort_index().plot(kind='bar', ax=axes[0], color='steelblue', edgecolor='white')
axes[0].set_title('Distribusi Rating (GoodBooks-10k)')
axes[0].set_xlabel('Rating (1-5)')
axes[0].set_ylabel('Jumlah')

books['average_rating'].plot(kind='hist', bins=20, ax=axes[1], color='coral', edgecolor='white')
axes[1].set_title('Distribusi Average Rating Buku')
axes[1].set_xlabel('Average Rating')

plt.tight_layout()
plt.savefig('rating_distribution.png', dpi=150)
plt.show()
print(f'Total ratings: {len(ratings):,}')
print(f'Rating mean: {ratings["rating"].mean():.2f}')
```

```python
# Sparsity matrix
n_users   = ratings['user_id'].nunique()
n_books   = ratings['book_id'].nunique()
n_ratings = len(ratings)
sparsity  = 1 - (n_ratings / (n_users * n_books))
print(f'Users:    {n_users:,}')
print(f'Books:    {n_books:,}')
print(f'Ratings:  {n_ratings:,}')
print(f'Sparsity: {sparsity:.4%}')
```

```python
# Missing values
print('=== Books Missing Values ===')
print(books[['book_id','isbn','title','authors','original_publication_year','image_url']].isnull().sum())
```

```python
# Top 10 penulis paling banyak buku
top_authors = books['authors'].value_counts().head(10)
top_authors.plot(kind='barh', figsize=(8, 4), color='teal', edgecolor='white')
plt.title('Top 10 Penulis (GoodBooks-10k)')
plt.xlabel('Jumlah Buku')
plt.tight_layout()
plt.savefig('top_authors.png', dpi=150)
plt.show()
```

```python
# Distribusi jumlah rating per user
user_activity = ratings['user_id'].value_counts()
plt.figure(figsize=(10, 4))
plt.hist(user_activity[user_activity <= 100], bins=50, color='steelblue', edgecolor='white')
plt.xlabel('Jumlah Rating per User')
plt.ylabel('Frekuensi')
plt.title('Distribusi Aktivitas User')
plt.axvline(3, color='red', linestyle='--', label='Threshold (3 ratings)')
plt.legend()
plt.tight_layout()
plt.savefig('user_activity.png', dpi=150)
plt.show()
print(f'Users dengan < 3 rating: {(user_activity < 3).sum():,} ({(user_activity < 3).mean():.1%})')
```
