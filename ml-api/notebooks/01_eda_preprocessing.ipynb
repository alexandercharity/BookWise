# BookWise — EDA & Preprocessing
Dataset: Book-Crossing (Kaggle)
https://www.kaggle.com/datasets/ruchi798/bookcrossing-dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DATA_DIR = '../../data'

books   = pd.read_csv(f'{DATA_DIR}/BX-Books.csv',        sep=';', encoding='latin-1', on_bad_lines='skip')
ratings = pd.read_csv(f'{DATA_DIR}/BX-Book-Ratings.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
users   = pd.read_csv(f'{DATA_DIR}/BX-Users.csv',        sep=';', encoding='latin-1', on_bad_lines='skip')

books.columns   = books.columns.str.strip().str.replace('"', '')
ratings.columns = ratings.columns.str.strip().str.replace('"', '')
users.columns   = users.columns.str.strip().str.replace('"', '')

print('Books shape:  ', books.shape)
print('Ratings shape:', ratings.shape)
print('Users shape:  ', users.shape)
```

```python
# Distribusi rating
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

ratings['Book-Rating'].value_counts().sort_index().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Distribusi Semua Rating')
axes[0].set_xlabel('Rating')

explicit = ratings[ratings['Book-Rating'] > 0]
explicit['Book-Rating'].value_counts().sort_index().plot(kind='bar', ax=axes[1], color='coral')
axes[1].set_title('Distribusi Rating Eksplisit (1-10)')
axes[1].set_xlabel('Rating')

plt.tight_layout()
plt.savefig('rating_distribution.png', dpi=150)
plt.show()
print(f'Implicit (0): {(ratings["Book-Rating"]==0).sum():,}')
print(f'Explicit (1-10): {(ratings["Book-Rating"]>0).sum():,}')
```

```python
# Sparsity matrix
n_users = ratings['User-ID'].nunique()
n_books = ratings['ISBN'].nunique()
n_ratings = len(ratings)
sparsity = 1 - (n_ratings / (n_users * n_books))
print(f'Users: {n_users:,}')
print(f'Books: {n_books:,}')
print(f'Ratings: {n_ratings:,}')
print(f'Sparsity: {sparsity:.4%}')
```

```python
# Missing values
print('=== Books Missing Values ===')
print(books.isnull().sum())
print('\n=== Users Missing Values ===')
print(users.isnull().sum())
```

```python
# Top 10 penulis paling banyak buku
top_authors = books['Book-Author'].value_counts().head(10)
top_authors.plot(kind='barh', figsize=(8, 4), color='teal')
plt.title('Top 10 Penulis')
plt.tight_layout()
plt.savefig('top_authors.png', dpi=150)
plt.show()
```
